In general, words can be used to represent a common class of feature. Word characteristics are also features. For example, capitalization matters: US versus us, White House versus white house. Part of speech and grammatical structure also add to textual features. Semantics can be a textual feature: buy versus purchase.

The mapping from textual data to real-valued vectors is called feature extraction. One of the simplest techniques to numerically represent text is Bag of Words (BOW). In BOW, we make a list of unique words in the text corpus called vocabulary. Then we can represent each sentence or document as a vector, with each word represented as 1 for presence and 0 for absence.

Another representation is to count the number of times each word appears in a document. The most popular approach is using the Term Frequency-Inverse Document Frequency (TF-IDF) technique.

```python
def cleanpunc(sentence): #function to clean the word of any punctuation or special characters
    cleaned = re.sub(r'[?|!|\'|"|#]',r'',sentence)
    cleaned = re.sub(r'[.|,|)|(|\|/]',r' ',cleaned)
    return  cleaned
```

```python
i=0
list_of_sent_train=[]
for sent in df['Summary2'].values:
    filtered_sentence=[]
    for w in sent.split():
        for cleaned_words in cleanpunc(w).split():
            if(cleaned_words.isalpha()):
                filtered_sentence.append(cleaned_words.lower())
            else:
                continue
    list_of_sent_train.append(filtered_sentence)
```

```python
w2v_model=gensim.models.Word2Vec(list_of_sent_train,vector_size=100, workers=4)
```

```python
import numpy as np
sent_vectors = [];
for sent in list_of_sent_train:
    sent_vec = np.zeros(100)
    cnt_words =0;
    for word in sent:
        try:
            vec = w2v_model.wv[word]
            sent_vec += vec
            cnt_words += 1
        except:
            pass
    sent_vec /= cnt_words
    sent_vectors.append(sent_vec)
sent_vectors = np.array(sent_vectors)
sent_vectors = np.nan_to_num(sent_vectors)
sent_vectors.shape
```

```python
num_clus = [x for x in range(3,11)]
num_clus
```

Choosing the best cluster using Elbow Method.

```python
squared_errors = []
for cluster in num_clus:
    kmeans = KMeans(n_clusters = cluster).fit(sent_vectors) # Train Cluster
    squared_errors.append(kmeans.inertia_) # Appending the squared loss obtained in the list

optimal_clusters = np.argmin(squared_errors) + 2 # As argmin return the index of minimum loss.
plt.plot(num_clus, squared_errors)
plt.title("Elbow Curve to find the no. of clusters.")
plt.xlabel("Number of clusters.")
plt.ylabel("Squared Loss.")
xy = (optimal_clusters, min(squared_errors))
plt.annotate('(%s, %s)' % xy, xy = xy, textcoords='data')
plt.show()

print ("The optimal number of clusters obtained is - ", optimal_clusters)
print ("The loss for optimal cluster is - ", min(squared_errors))
```

```python
import pandas as pd
from transformers import BertTokenizer, BertForSequenceClassification
from itertools import combinations
from collections import defaultdict
import torch
import nltk
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.corpus import stopwords

# Download NLTK resources
nltk.download('stopwords')
nltk.download('punkt')

key_entities = ["israel", "palestine", "gaza", "west bank", "hamas", "israeli", "palestinian", "biden", "violence", "conflict", "two state, iran, fighting, death, fire"]

# Load stop words
stop_words = set(stopwords.words('english'))

# Load pre-trained BERT model and tokenizer
model_name = 'bert-base-uncased'
tokenizer = BertTokenizer.from_pretrained(model_name)
model = BertForSequenceClassification.from_pretrained(model_name)

# Function to preprocess text
def preprocess_text(text):
    sentences = sent_tokenize(text)
    words = [word_tokenize(sentence) for sentence in sentences]
    words = [word.lower() for sentence_words in words for word in sentence_words if word.isalpha() and word.lower() not in stop_words]
    return ' '.join(words)  # Combine the list of words into a single string

# Function to analyze sentiment of text using BERT
def analyze_sentiment(text):
    preprocessed_text = preprocess_text(text)

    # Tokenize and convert to tensor
    inputs = tokenizer(preprocessed_text, return_tensors='pt', truncation=True)

    # Forward pass
    outputs = model(**inputs)

    # Use the logits for sentiment analysis
    logits = outputs.logits
    probabilities = torch.nn.functional.softmax(logits, dim=1)

    # Get the predicted sentiment (positive/negative)
    predicted_sentiment = torch.argmax(probabilities, dim=1).item()

    print(f"Text: {text}")
    print(f"Preprocessed Text: {preprocessed_text}")
    print(f"Predicted Sentiment: {'positive' if predicted_sentiment == 1 else 'negative'}")

    return predicted_sentiment

# Apply sentiment analysis to each sentence in the 'summary' column
df['sentence_sentiment_bert'] = df['summary'].apply(lambda x: [analyze_sentiment(sentence) for sentence in sent_tokenize(x)])
```

```python
import pandas as pd
from textblob import TextBlob
from itertools import combinations
from collections import defaultdict
import nltk
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.corpus import stopwords

# Assuming you have a DataFrame named df with columns: title, date, summary, excerpt, sentiment_scores

# Download NLTK resources
nltk.download('stopwords')
nltk.download('punkt')

# Load stop words
stop_words = set(stopwords.words('english'))

# Function to preprocess text
def preprocess_text(text):
    sentences = sent_tokenize(text)
    words = [word_tokenize(sentence) for sentence in sentences]
    words = [word.lower() for sentence_words in words for word in sentence_words if word.isalpha() and word.lower() not in stop_words]
    return words

# Function to analyze sentiment of sentences using TextBlob
def analyze_sentiment(sentence):
    preprocessed_sentence = preprocess_text(sentence)
    analysis = TextBlob(preprocessed_sentence)
    print(f"Sentence: {sentence}")
    print(f"Preprocessed Sentence: {preprocessed_sentence}")
    print(f"Raw Sentiment Score: {analysis.sentiment}")
    return analysis.sentiment.polarity

# Apply sentiment analysis to each relevant sentence
df['sentence_sentiment'] = df['summary'].apply(lambda x: [analyze_sentiment(sentence) for sentence in x])

# Task 3: Create a Co-occurrence Matrix

# Function to update co-occurrence matrix based on entities in sentences
def update_cooccurrence_matrix(entities, cooccurrence_matrix):
    for pair in combinations(entities, 2):
        cooccurrence_matrix[pair] += 1

# Initialize co-occurrence matrix
cooccurrence_matrix = defaultdict(int)

# Apply the update_cooccurrence_matrix function to each row
df['relevant_entities'] = df['relevant_sentences'].apply(lambda x: list(set([entity for sentence in x for entity in key_entities])))
df['relevant_entities'].apply(lambda x: update_cooccurrence_matrix(x, cooccurrence_matrix))

# Task 4: Analyze Patterns

# Convert the co-occurrence matrix to a DataFrame
cooccurrence_df = pd.DataFrame(list(cooccurrence_matrix.items()), columns=['Entity Pair', 'Co-occurrence Count'])

# Display the co-occurrence DataFrame
print(cooccurrence_df)
```
